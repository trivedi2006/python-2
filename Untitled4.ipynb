```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randint(1,100,(20,5)),columns=["A","B","C","D","E"])
```

```python
df
```

```python
df.apply(sum)
```

```python
df["total"] = df.apply(sum,axis=1)
```

```python
df.apply("mean")
```

```python
df.apply(lambda x:x["A"]+x["B"]+x["C"],axis=1)
```

```python
df
```

```python
def grade(x):
    if x>300:
        return "A"
    if x>250:
        return "B"
    if x>200:
        return "C"
    return "F"

         
       
```

```python
df["grade"] = df["total"].apply(grade)
df
```

```python
df["result"] = df.apply(lambda x:"Fail" if (x["A"]<35) or (x["B"]<35) or (x["C"]<35)or (x["D"]<35) or (x["E"]<35) else "Pass",axis = 1 )
```
