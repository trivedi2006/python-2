```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("social_media_engagement.csv")
df
```

```python
def risk(x):
    if x["likes"]>500 and x["comments"]<100:
        return "High Risk"
    if  x["likes"]>2000:
        return "Medium"
    if x["shares"]>1000:
        return "Low"
    
        
```

```python
df["risk"] = df.apply(risk,axis=1)
```
