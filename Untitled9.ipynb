```python
import numpy as np
import pandas as pd
import plotly.express as px
import plotly.io as pio
pio.renderers.default="notebook"
```

```python
df = pd .DataFrame({"dep":["hr","it","hr","it","hr"],"age":[20,25,30,22,10],"sal":[100,500,200,300,100],"sex":["f","m","m","f","f"]})
```

```python
px.histogram(data_frame=df,x = "age",color = "sex",pattern_shape = "dep")
```

```python
px.histogram(data_frame=df,x = "dep",y = "sal")
```

```python
px.histogram(data_frame=df,x = "dep",y = "sal",histfunc="max")
```

```python
px.histogram(data_frame=df,x = "dep",y = "sal",histfunc="avg")
```

```python
px.histogram(data_frame=df,x = "dep",y = "sal",histfunc="max",color = "sex")
```

```python
import networkx as nx
```

```python
g = nx.Graph()
g.add_node(1)
g.add_node(2)
g.add_node(3)
g.add_node(4)
g.add_edge(1,2)
g.add_edge(2,3)
g.add_edge(3,4)
g.add_edge(4,1)
nx.draw(g,with_labels = True,node_size=1000)
```

```python
g = nx.DiGraph()
g.add_nodes_from([1,2,3])
g.add_edges_from([(1,2),(2,3),(3,1),(1,1)])
nx.draw(g,with_labels = True,node_size=1000)
```
