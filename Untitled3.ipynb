```python
import seaborn as sns
```

```python
sns.boxenplot(x = "mpg",data=auto)
```

```python
q1 = auto.mpg.quantile(0.25)
q1
```

```python
q3 = auto.mpg.quantile(0.75)
q3
```

```python
auto.mpg.describe()
```

```python
IQR = q3-q1
IQR
```

```python
Hl = q3+1.5*IQR
LL = q1-1.5*IQR
Hl,LL
```

```python
wo = auto[(auto["mpg"]<Hl) &  (auto["mpg"]>LL)]
wo
```

```python
sns.boxenplot(x = "mpg",data=wo)
```
