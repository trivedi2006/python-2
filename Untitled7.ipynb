```python
df1 = pd.DataFrame({"name":["vha","psp","vby","mss","dps"],"sub":["python","fsd","coa","dm","toc"],"id":[3,2,1,4,6]})
```

```python
df2 = pd.DataFrame({"id":[1,2,3,4,5],"mark":[9,8,10,8,9]})
```

```python
pd.merge(df1,df2,on="id")
```

```python
pd.merge(df1,df2,on="id",how="left")
```

```python
pd.merge(df1,df2,how="cross")
```

```python
df1 = pd.DataFrame({"name":["vha","psp","vby","mss","dps"],"sub":["python","fsd","coa","dm","toc"],"in":[3,2,1,4,6]})
```

```python
df2 = pd.DataFrame({"id":[1,2,3,4,5],"mark":[9,8,10,8,9],})
```

```python
df1.merge(df2,left_on="name",right_on="fac",how="outer",suffixes = ("_b1_sem3","_b1_sem4"))
```
